```python
df_credit_grouped = df_credit.groupby('ID')
```

```python
df_credit_pivot = df_credit.pivot(index ='ID', columns = 'MONTHS_BALANCE',values = 'STATUS')
```

```python
df_credit_pivot['opening_month'] = df_credit_grouped['MONTHS_BALANCE'].min()
```

```python
df_credit_pivot['closing_month'] = df_credit_grouped['MONTHS_BALANCE'].max()
```

```python
df_credit_pivot['ID'] = df_credit_pivot.index
```

```python
df_credit_pivot = df_credit_pivot[['ID','opening_month','closing_month']]
```

```python
df_credit_pivot['window'] = df_credit_pivot['closing_month'] - df_credit_pivot['opening_month']
```

```python
df_credit_pivot
```

```python
df_credit_pivot.reset_index(drop = True, inplace = True)
```

```python
df_credit_pivot.columns.to_flat_index()
```

```python
df_credit = pd.merge(df_credit,df_credit_pivot, on ='ID', how = 'left')
```

```python
df_credit_copy = df_credit.copy()
```

```python
df_credit.STATUS.unique()
```

```python
df_credit  = df_credit[df_credit['window'] > 20]
```

```python
df_credit['status'] = np.where((df_credit['STATUS'] == '2') | (df_credit['STATUS'] == '3' )| (df_credit['STATUS'] == '4' )| (df_credit['STATUS'] == '5'), 1, 0)
```

```python
df_credit['status'] = df_credit['status'].astype(np.int8)
```

```python
df_credit['month_on_book'] = df_credit['MONTHS_BALANCE'] - df_credit['opening_month']
```

```python
df_credit
```

```python
df_credit.sort_values(by=['ID','month_on_book'], inplace = True)
```

```python
new = df_credit_pivot.groupby(['opening_month']).agg({'ID':['count']})
```

```python
new.reset_index(inplace = True)
```

```python
new.columns = ['opening_month','ID_sum']
```

```python
vintage = df_credit.groupby(['opening_month','month_on_book']).agg({'ID':['count']})
```

```python
vintage.reset_index(inplace = True)
```

```python
vintage.columns = ['opening_month','month_on_book','count']
```

```python
vintage['due_count'] = np.nan
```

```python
vintage = vintage[['opening_month','month_on_book','due_count']]
```

```python
vintage = pd.merge(vintage, new, on = ['opening_month'], how = 'left')
```

```python
# vintage
```

```python
for j in range(-60,0):
    n = []
    for i in range(0,60):
        due = list(df_credit[(df_credit['status'] == 1) & (df_credit['month_on_book'] == i) & (df_credit['opening_month'] == j)]['ID'])
        n.extend(due)
        vintage.loc[(vintage['month_on_book'] == i) & (vintage['opening_month'] == j), 'due_count'] = len(set(n))
              
vintage['sta_rate']  = vintage['due_count'] / vintage['ID_sum']
```

```python
vintage_wide = vintage.pivot(index = 'opening_month',
                             columns = 'month_on_book',
                             values = 'sta_rate')
vintage_wide
```

```python
def calculate_rate(df_credit_pivot, command): 

    df_credit_copy['status'] = None
    exec(command)
    sumagg = df_credit_copy.groupby('ID')['status'].agg(sum)
    df_credit_pivot = pd.merge(df_credit_pivot, sumagg, on = 'ID', how = 'left')
    df_credit_pivot.loc[df_credit_pivot['status'] > 1, 'status'] = 1
    rate = df_credit_pivot['status'].sum() / len(df_credit_pivot)

    return round(rate, 5)
```

```python
command = "df_credit_copy.loc[(df_credit_copy['STATUS'] == '0') | (df_credit_copy['STATUS'] == '1') | (df_credit_copy['STATUS'] == '2') | (df_credit_copy['STATUS'] == '3' )| (df_credit_copy['STATUS'] == '4' )| (df_credit_copy['STATUS'] == '5'), 'status'] = 1"   
morethan1 = calculate_rate(df_credit_pivot, command)
```
